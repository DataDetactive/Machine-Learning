```python
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
```

```python
#simple data & showing the corrolation
xs = np.array([1,2,3,4,5,6],dtype=np.float64)
ys = np.array([5,4,6,5,6,7],dtype=np.float64)
plt.plot(xs,ys)
plt.show()
plt.scatter(xs,ys)
plt.show()
```

```python
#the function for the regression
def best_fit_solpe_intercept(xs,ys):
    m = ( (mean(xs) * mean(ys)) - mean(xs*ys) )/ ( (mean(xs)*mean(xs)) - mean(xs*xs) )
    b = ( (mean(ys) ) ) - ( (m) * (mean(xs)) )
    return m,b
```

```python

m , b = best_fit_solpe_intercept(xs,ys)
print("slope is =", m, "\n" ,"Y intercept is =",b)
```

```python
#the prediction of Y
regression_line = [(m*x)+b for x in xs]

print(regression_line)
```

```python
#the best fit line in that sample
plt.scatter(xs,ys)
plt.plot(xs,regression_line,'r')
plt.show()
```

```python
# the accuracy is equavlient to coefficient of determination in linear regression so the below how we calculate the accuracy of that sample 
```

```python
def squared_error(ys_orginal,ys_line):
    return sum((ys_line-ys_orginal)**2)
```

```python

def coefficient_of_determination(ys_original,ys_thepredict):
    y_mean_orginals_list=[ mean(ys_original) for y in ys_original]
    squared_error_Y_predict= squared_error(ys_original,ys_thepredict)
    squared_error_Y_orginal= squared_error( ys_original , y_mean_orginals_list)
    return   1- (squared_error_Y_predict / squared_error_Y_orginal )
```

```python
r_squared = coefficient_of_determination(ys,regression_line)
```
